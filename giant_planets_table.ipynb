```python
from astroquery.simbad import Simbad
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', None)
import wget
import os
```

```python
pdnfinal=pd.read_csv("exoplanet.eu_catalog_simbad.csv")
print(pdnfinal.columns)

pdnfinal=pdnfinal[pdnfinal['star_distance']< 20]
pdnfinal=pdnfinal[pdnfinal['orbital_period']< 10]

#From Jupiter to Earth units
pdnfinal['mass']=pdnfinal['mass']*3.1782838*1e2
pdnfinal['mass_sini']=pdnfinal['mass_sini']*3.1782838*1e2
pdnfinal['radius']=pdnfinal['radius']*11.209

#pdnfinal=pdnfinal[pdnfinal['mass']> 17] #neptune mass
pdnfinal=pdnfinal[pdnfinal['radius']> 2]

pdnfinal=pdnfinal[['# name','simbad_id','ra', 'dec','mass','mass_sini','radius','orbital_period', 'semi_major_axis','eccentricity','star_distance','star_mass','star_radius', 'star_sp_type','star_teff','star_magnetic_field','detection_type',
       'mass_detection_type', 'radius_detection_type']]

pdnfinal["bfield_star(gauss)"]=100*np.pi
pdnfinal["p_rot(days)"]=10*np.pi

pdnfinal=pdnfinal.rename(columns={"# name":"planet_name","simbad_id":"star_name","star_distance":"d_star(pc)",
                                 "star_radius":"radius_star(r_sun)","star_mass":"mass_star(m_sun)","orbital_period":"p_orb(days)",
                                 "radius":"radius_planet(r_earth)","mass":"mass_planet(m_earth)","semi_major_axis":"a(au)"})

pdnfinal.to_csv('GMRT-giant_planets_proposal.csv')
```
